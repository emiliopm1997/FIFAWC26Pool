```python
from pathlib import Path
import sqlite3

MAIN_PATH = Path().cwd().parent
SCHEMA_PATH = MAIN_PATH / "templates" / "schema.sql"
DB_PATH = MAIN_PATH / "data" / "fifa_wc_2026_pool.db"
```

### Create DB and structure

This only works if the db file doesn't exist.

```python
if not DB_PATH.exists():
    connection = sqlite3.connect(DB_PATH)

    with open(SCHEMA_PATH, "r") as f:
        schema_sql = f.read()

    connection.executescript(schema_sql)
    print("Database created and schema applied successfully.")

    connection.close()
```
